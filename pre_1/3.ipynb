### \#\# `defaultdict`의 핵심 개념과 목적

제공해주신 코드는 텍스트 내 단어의 빈도를 계산하는 전형적인 예시이며, `defaultdict`의 장점을 명확히 보여줍니다.

1.  **KeyError 방지**

      * 일반 딕셔너리(`dict`)는 존재하지 않는 키를 조회하거나 값을 수정하려 하면 `KeyError`가 발생한다.
      * `defaultdict`는 생성 시 지정한 '기본값 생성 함수(default factory)'를 호출하여 해당 키와 기본값을 자동으로 만들어주므로 오류가 발생하지 않는다. 예시의 `lambda: 0`이 바로 이 함수이며, 존재하지 않는 키에 접근 시 숫자 `0`을 기본값으로 만들어준다.

2.  **코드 간소화**

      * `defaultdict`를 사용하지 않으면, 키의 존재 여부를 매번 확인하는 코드가 필요하다.
          * **일반 dict 사용 시:**
            ```python
            word_count = {}
            for word in text:
                if word not in word_count:
                    word_count[word] = 0 # 키가 없으면 초기화
                word_count[word] += 1
            ```
      * `defaultdict`를 사용하면 이러한 조건문 없이 바로 값을 조작할 수 있어 코드가 훨씬 짧고 직관적이 된다. 예시의 `word_count[word] += 1` 한 줄로 모든 처리가 끝난다.

3.  **다양한 활용성**

      * 기본값은 숫자에 국한되지 않는다. 예를 들어 `defaultdict(list)`로 설정하면, 존재하지 않는 키에 접근 시 빈 리스트(`[]`)가 기본값으로 생성된다. 이를 통해 특정 키에 해당하는 값들을 리스트로 묶는(grouping) 작업을 쉽게 처리할 수 있다.
      * **출처:** 이 기능은 Python 공식 문서의 `collections.defaultdict` 항목에 명시된 핵심 용도이다.

-----

### \#\# 예시 코드 상세 분석

  * **`word_count = defaultdict(lambda: 0)`**

      * `defaultdict` 객체를 생성한다.
      * `lambda: 0`은 이름 없는 간단한 함수로, 호출될 때마다 숫자 `0`을 반환한다.
      * 이 설정은 `word_count` 딕셔너리에서 어떤 단어(키)를 처음 만났을 때, 그 단어의 빈도수(값)를 `0`으로 자동 설정하겠다는 의미이다.
      * (참고: 코드 상단에는 `OrderedDict`가 import 되었지만, 실제 객체 생성은 `defaultdict`를 사용하고 있습니다. `from collections import defaultdict`가 누락된 것으로 보입니다.)

  * **`for word in text:`**

      * `text`라는 변수(문자열 혹은 리스트로 추정)에 포함된 모든 단어를 하나씩 순회한다.

  * **`word_count[word] += 1`**

      * **이 부분이 `defaultdict`의 핵심이다.**
      * **만약 `word`가 `word_count`에 처음 등장하는 단어라면:**
        1.  `word_count[word]`에 접근하는 순간, 키가 없음을 `defaultdict`가 인지한다.
        2.  미리 지정된 `lambda: 0`을 호출하여 값으로 `0`을 생성하고 `word_count[word] = 0`을 자동으로 실행한다.
        3.  그 후에 `+= 1` 연산을 수행하여 최종적으로 `word_count[word]`는 `1`이 된다.
      * **만약 `word`가 이미 `word_count`에 존재하는 단어라면:**
        1.  기존 값에 `1`을 더한다.

  * **`for i, v in OrderedDict(sorted(...)).items():`**

      * 이 부분은 `defaultdict`의 역할이 끝난 후, 결과를 정렬하여 출력하는 부분이다.
      * `word_count.items()`: `(단어, 빈도수)` 형태의 쌍들을 가져온다.
      * `sorted(..., key=lambda t: t[1], reverse=True)`: 각 쌍(`t`)에서 값(빈도수, `t[1]`)을 기준으로 내림차순(`reverse=True`) 정렬한다.
      * `OrderedDict(...)`: 정렬된 순서를 기억하는 특수한 딕셔너리(`OrderedDict`)로 변환한다.
      * `for i, v in ... .items()`: 정렬된 딕셔너리에서 키(`i`)와 값(`v`)을 하나씩 꺼내 `print(i, v)`로 출력한다.

````python

text = 'hello wrodl! this is a sample sentence to test for defaultdict type of python'.split()

word_count = defaultdict(lambda: 0)
for word in text:
    word_count[word] += 1
for i, v in OrderedDict(sorted(word_count.items(), key=lambda t: t[1], reverse=True)).items():
    print(i, v)
````

물론입니다. Jarvis입니다. 바로 결론부터 말씀드리겠습니다.

`key=lambda t: t[1]` 코드는 `sorted` 함수에게 **"정렬 대상이 되는 각 항목(t)에서, 1번 인덱스에 있는 값(t[1])을 기준으로 삼아 정렬하라"** 는 구체적인 '정렬 규칙'을 알려주는 역할을 합니다.

### \#\# `lambda t: t[1]` 상세 분석

이 코드를 이해하기 위해 세 부분으로 나누어 설명하겠습니다.

#### 1\. `lambda`: 이름 없는 한 줄짜리 함수

  * **의미**: `lambda`는 파이썬에서 사용하는 키워드로, 작고 간단한 '익명 함수(anonymous function)'를 만들 때 사용한다. `def` 키워드로 정식 함수를 정의하기에는 너무 간단한 로직을 한 줄로 표현할 때 유용하다.
  * **구조**: `lambda 인자: 표현식` 형태로 작성된다. 인자를 받아서 표현식을 실행한 결과를 반환한다.
  * **비교**: `lambda t: t[1]`은 아래의 일반 함수와 완전히 동일한 기능을 한다.
    ```python
    def get_second_element(t):
        return t[1]

    # sorted(..., key=get_second_element) 와 동일함
    ```
    이처럼 간단한 기능을 위해 별도의 함수를 정의하는 대신, 필요할 때 즉석에서 만들어 사용하는 것이 `lambda`이다.

#### 2\. `t`: 처리될 각 데이터 항목을 가리키는 변수

  * **의미**: `t`는 `lambda` 함수가 전달받는 인자(parameter)의 이름이다. 이 이름은 프로그래머가 임의로 정할 수 있으며, `item`, `x` 등 다른 이름을 사용해도 무방하다.
  * **역할**: `sorted` 함수는 정렬할 리스트(`word_count.items()`)의 모든 항목을 하나씩 순회하며 이 `lambda` 함수에 전달한다. 이때 `t`는 순서대로 각 항목을 가리키게 된다.
  * **실제 값**: 이전 예제에서 `word_count.items()`는 `[('a', 12), ('to', 10), ('and', 9), ...]` 와 같은 형태의 리스트이다. 따라서 `sorted` 함수가 동작하는 동안, 변수 `t`는 다음과 같이 변한다.
      * 첫 번째 순회에서 `t`는 `('a', 12)` 이다.
      * 두 번째 순회에서 `t`는 `('to', 10)` 이다.
      * 세 번째 순회에서 `t`는 `('and', 9)` 이다.
      * ... (리스트 끝까지 반복)

#### 3\. `t[1]`: 정렬의 기준이 되는 실제 값

  * **의미**: 파이썬에서 `[ ]`는 인덱싱(indexing)을 의미하며, 시퀀스(리스트, 튜플 등)의 특정 위치에 있는 요소에 접근하는 방법이다. 인덱스는 `0`부터 시작한다.
  * **역할**: `t`가 `('a', 12)`와 같은 튜플(tuple)이므로, 각 인덱스는 다음을 가리킨다.
      * `t[0]`: 0번 인덱스, 즉 첫 번째 요소인 단어 `'a'`
      * **`t[1]`: 1번 인덱스, 즉 두 번째 요소인 숫자 `12`**
  * **결론**: 따라서 `t[1]`은 각 튜플에서 **'단어의 빈도수(count)'** 를 정확히 추출해낸다. `sorted` 함수는 이 `lambda` 함수가 반환하는 값들(12, 10, 9, 9, 8...)을 비교하여 전체 데이터를 정렬하게 된다.

````python
c = Counter (a = 4, b= 2, c= 0, d = -1)
d = Counter (a = 1, b = 2, c = 3, d = 4)

print(c) 
print(d) 

print(c.subtract(d))
print(d.subtract(c))
print(c | d)
print(c & d)
print(c - d)

````

**`namedtuple`은 튜플(tuple)의 각 원소에 이름을 부여하여, 인덱스 번호뿐만 아니라 `.`을 이용한 이름으로도 데이터에 접근할 수 있게 해주는 자료구조**입니다. 이를 통해 코드의 가독성을 획기적으로 높이는 것이 주된 목적입니다.

### ## `namedtuple`의 핵심 목적과 용도

제공해주신 코드는 CSV 파일에서 읽어온 데이터를 `namedtuple`로 변환하여 처리하는 좋은 예시입니다. `namedtuple`을 사용하는 이유는 다음과 같습니다.

1.  **코드 가독성 향상 (Clarity)**
    * 일반 튜플이나 리스트는 데이터에 접근할 때 `student[4]` 와 같이 인덱스를 사용해야 한다. 이 숫자 `4`가 무엇을 의미하는지 알려면 데이터 구조를 따로 찾아봐야 한다.
    * `namedtuple`을 사용하면 `student.first_name` 처럼 의미 있는 이름으로 데이터에 접근할 수 있다. 이는 코드를 읽는 것만으로도 데이터의 의미를 즉시 파악할 수 있게 하여 유지보수를 쉽게 만든다.

2.  **튜플의 고유 특성 유지 (Efficiency & Immutability)**
    * `namedtuple`은 튜플을 기반으로 하므로, 일반 튜플처럼 메모리를 적게 사용하고 속도가 빠르다.
    * 또한, 생성된 후에는 값을 변경할 수 없는 '불변성(Immutability)'을 그대로 유지한다. 따라서 실수로 데이터가 변경될 염려가 없는, 안정적인 데이터 레코드(record)를 저장하기에 적합하다.

3.  **경량 객체(Lightweight Object) 역할**
    * 간단한 데이터 구조를 위해 매번 `class`를 정의하는 것은 번거로울 수 있다. `namedtuple`은 별도의 클래스 선언 없이, 한 줄의 코드로 간단하게 데이터 필드를 갖는 새로운 타입을 생성할 수 있는 편리한 대안이다.
    * **출처**: 이 개념은 Python 공식 문서의 `collections.namedtuple` 항목에서 "memory-efficient alternative to a regular class"로 소개되고 있다.

### ## 예시 코드 상세 분석


코드의 흐름은 **'CSV 파일 읽기 → 일반 리스트로 저장 → `namedtuple` 타입 정의 → `namedtuple` 객체로 변환하여 새 리스트에 저장'** 순서로 진행됩니다.

* **`Student = namedtuple('Student', " ".join(coloumns))`**
    * `namedtuple`을 이용해 새로운 클래스(타입)를 만드는 과정이다.
    * 첫 번째 인자인 `'Student'`는 생성될 클래스의 이름이 된다.
    * 두 번째 인자는 필드(field)들의 이름을 공백으로 구분한 문자열이다. `coloumns` 리스트를 `" ".join()`으로 합쳐 `"user_id integration_id ..."` 와 같은 문자열로 만들었다.
    * 이 한 줄의 코드로 `user_id`, `integration_id` 등의 속성을 갖는 `Student` 라는 새로운 자료형이 만들어진다.

* **`for row in student_list:`**
    * CSV 파일에서 읽어온 데이터가 담긴 `student_list`를 한 줄씩 순회한다. 이때 각 `row`는 `['101', '...', 'John', 'Doe', ...]` 와 같은 형태의 일반 리스트이다.

* **`student = Student(*row)`**
    * **이 부분이 `namedtuple`을 활용하는 핵심이다.**
    * `*row`: `*`(Asterisk)는 '언패킹(Unpacking)' 연산자이다. 리스트인 `row`의 모든 원소를 풀어서 `Student` 생성자에 각각의 인자로 전달한다.
    * 즉, `Student('101', '...', 'John', 'Doe', ...)` 와 같이 코드가 실행된다.
    * 결과적으로 `student` 변수에는 각 데이터가 이름에 매핑된 `namedtuple` 객체가 저장된다.

* **`print(student_namedtuple_list[0].full_name)`**
    * `namedtuple`의 가장 큰 장점을 보여주는 마지막 줄이다.
    * `student_namedtuple_list[0]`: 리스트의 첫 번째 `Student` 객체를 가져온다.
    * `.full_name`: 인덱스 번호(`[4]` 등) 대신 `.이름` 형식으로 `full_name` 필드의 값에 직접 접근한다.
    * **(주의)**: 다만, 제시된 코드에서는 `coloumns` 리스트에 `full_name`이 정의되어 있지 않아 이 코드는 실제로 실행하면 **오류(AttributeError)**가 발생한다. `student_namedtuple_list[0].first_name` 과 같이 정의된 필드 이름을 사용해야 올바르게 동작한다.

````python

student_list = [
    "1 0001 1234 han ryu",
    "2 0002 1234 han ryu1",
    "3 0003 1234 han ryu2",
    "4 0004 1234 han ryu3",
    "5 0005 1234 han ryu4",
]

columns = ["id", "login_id", "password", "first_name", "last_name"]
Student = namedtuple('Student', " ".join(columns)) # 타입 생성, 클래스 
student_namedtuple_list = []
for row in student_list:
    student = Student(*row.split()) # asterisk 뭔지 알아야함 unpack 기능이라고 했음
    student_namedtuple_list.append(student)

print(student_namedtuple_list)
print(student_namedtuple_list[-1].last_name)
````

핵심은 `*`가 **'이미 분리된 항목들의 묶음(iterable)'**을 풀어주는 것이지, **'하나의 덩어리를 알아서 의미 단위로 분리'**해주는 것이 아니라는 점입니다.

### ## `*` 연산자의 역할: 포장 풀기

`*` 연산자를 '선물 상자의 포장을 푸는 행위'에 비유할 수 있습니다.

1.  **잘못된 경우: `*` 와 문자열**
    * `row = "1 0001 1234 han ryu"`
    * 이것은 마치 '과일맛사탕'이라는 글씨가 쓰인 **하나의 긴 젤리**와 같습니다.
    * `*row`는 이 젤리를 성분 단위인 **문자(`'1'`, `' '`, `'0'`, ...)** 로 모두 분해해버립니다. 함수는 과일 사탕을 원했는데, 알 수 없는 가루들을 받은 셈입니다.

2.  **올바른 경우: `*` 와 리스트**
    * `row.split() = ['1', '0001', '1234', 'han', 'ryu']`
    * 이것은 '사과', '포도', '딸기' 사탕이 각각 들어있는 **선물 상자**와 같습니다. 각 항목이 이미 의미 있는 단위로 분리되어 있습니다.
    * `*row.split()`는 이 **선물 상자의 포장을 풀어** 내용물인 `'1'`, `'0001'`, `'1234'`, `'han'`, `'ryu'`를 각각 꺼내어 함수에 하나씩 전달합니다. 함수가 정확히 원하던 방식입니다.

따라서 `Student(*row.split())` 코드는 "**`row`를 공백으로 잘라 선물 상자(`리스트`)를 만들고, 그 포장(`*`)을 풀어 내용물들을 `Student` 함수에 각각 전달해줘**" 라는 의미로 완벽하게 동작하게 됩니다.

## 파이썬 자료구조 및 복사 메커니즘 핵심 정리
오늘 대화의 핵심은 "데이터를 어떻게 담고, 어떻게 복사하는가?" 였습니다. 특히 복사 방식은 컨테이너 안의 요소가 어떤 타입인가에 따라 결과가 완전히 달라진다는 점이 중요합니다.

### 1. defaultdict 와 namedtuple: 코드의 가독성과 편의성

#### defaultdict: KeyError 방지용 딕셔너리

- 목적: 존재하지 않는 키(key)에 접근할 때, 미리 지정한 기본값을 자동으로 생성하여 오류를 방지하고 코드를 간결하게 만든다.
- 핵심: 'if key in my_dict:' 와 같은 불필요한 확인 절차를 생략하게 해준다.

#### namedtuple: 이름으로 접근하는 튜플

- 목적: 튜플의 각 원소에 이름을 부여하여, 인덱스([0])가 아닌 이름(.name)으로 데이터에 접근하게 해 코드의 가독성을 극대화한다.
- 특징: 튜플의 특성(불변성, 효율성)을 유지하면서 간단한 데이터 객체처럼 사용할 수 있다.

### 2. * (Asterisk): 묶음을 푸는 연산자

- 역할: 리스트나 튜플 같은 **'항목들의 묶음(iterable)'**을 풀어헤쳐, 각각을 별개의 인자(argument)로 전달한다.
- 주의: *는 의미 단위로 분리하는 기능이 없다. 따라서 * 'a b c'는 'a', 'b', 'c'가 아닌 'a', ' ', 'b', ' ', 'c' 로 분해된다. 반드시 .split() 등으로 의미 있는 묶음(리스트 등)을 만든 후 사용해야 한다.

### 3. 얕은 복사 vs 깊은 복사: 가장 중요한 개념 🧠

#### 얕은 복사 (Shallow Copy)

- 정의: 객체의 '껍데기(컨테이너)'만 복사하는 방식. 내부 요소들은 원본과 동일한 메모리 주소를 공유한다.
- 예시: 리스트 슬라이싱(new = old[:]), list(), 리스트 컴프리헨션, copy.copy()

#### 깊은 복사 (Deep Copy)

- 정의: 껍데기뿐만 아니라, 내부의 모든 요소들까지 재귀적으로 전부 복제하여 완전히 독립적인 객체를 만드는 방식.
- 예시: copy.deepcopy() (반드시 import copy 필요)

### 4. 얕은 복사의 결과가 달라지는 이유: 요소의 '변경 가능성'

- 얕은 복사를 했을 때, 그 결과는 전적으로 **내부 요소가 변경 가능한지(Mutable) 아닌지(Immutable)**에 달려 있습니다.

    - Case 1: 요소가 변경 불가능 (Immutable - str, int, tuple)
        - 이 타입들은 수정이 불가능하므로, 값을 바꾸는 연산(= 또는 +=)은 항상 새로운 객체를 만들어 재할당하는 방식으로 동작합니다.
        - 따라서 원본의 요소를 바꿔도 복사본은 옛날 요소를 그대로 참조하므로, 결과적으로 깊은 복사처럼 동작합니다.
    - Case 2: 요소가 변경 가능 (Mutable - list, dict)
        - 이 타입들은 그 자리에서 수정이 가능합니다.
        - 원본과 복사본이 동일한 내부 객체를 공유하므로, 복사본의 내부 요소를 수정하면 원본까지 함께 변경됩니다. 이것이 얕은 복사의 전형적인 특징입니다.

````python
for i, item in enumerate(listed_items):
    print(f'{i}: {item}')

new_list = list(enumerate(listed_items))

print(new_list)

new_dict = {i : j for i, j in enumerate(listed_items)} # dict 로 형변환
print(type(new_dict))
print(new_dict)
````

````python
def f(x: int, y: int) -> int:
    return x + y

f_lamda = lambda x, y: x + y  # 권장 안함 이유는? 하단 확인 

print(f(1, 5))
print(f_lamda(1, 5))
````

`lambda`가 권장되지 않는다는 것은 **'복잡한 작업'이나 '이름이 필요한 함수'를 `lambda`로 만드는 경우에 한정**된 이야기입니다. `lambda` 자체는 파이썬의 강력한 기능이며, 적절한 용도로 사용할 때는 매우 권장됩니다. 복잡한 `lambda`를 권장하지 않는 이유는 주로 **가독성과 디버깅의 어려움** 때문입니다.

-----

### 1. 가독성 저하 (Readability Counts)

파이썬의 핵심 철학 중 하나는 "가독성은 중요하다(Readability counts)"입니다. `lambda`는 한 줄로 함수를 표현해야 하므로, 로직이 조금만 복잡해져도 코드를 이해하기가 매우 어려워집니다.

  * **나쁜 예시 (복잡한 `lambda`)**:

    ```python
    # x가 짝수이면 제곱하고, 홀수이면 세 제곱하는 람다
    # 한 눈에 이해하기 어렵다.
    complex_lambda = lambda x: x**2 if x % 2 == 0 else x**3
    ```

  * **좋은 예시 (`def` 함수)**:

    ```python
    # 이름이 있고, 로직이 명확히 보인다.
    def calculate_power(x):
        if x % 2 == 0:
            return x**2  # 짝수는 제곱
        else:
            return x**3  # 홀수는 세 제곱
    ```

    `def`를 사용한 함수는 이름이 있어 **함수의 의도**를 명확히 알 수 있고, 여러 줄에 걸쳐 로직을 표현하므로 훨씬 이해하기 쉽습니다.

-----

### 2. 디버깅의 어려움 (Debugging Hell)

`lambda`는 '익명 함수(anonymous function)'입니다. 이름이 없기 때문에 프로그램 실행 중 오류가 발생했을 때, 에러 메시지에 어떤 함수에서 문제가 생겼는지 명확히 표시되지 않습니다.

  * **`lambda` 오류 메시지**: `TypeError in <lambda>`
      * 수많은 `lambda` 함수 중 어느 것에서 오류가 발생했는지 알기 어렵습니다.
  * **`def` 함수 오류 메시지**: `TypeError in calculate_power`
      * `calculate_power`라는 이름이 명확히 표시되므로, 문제의 원인을 즉시 찾을 수 있습니다.

-----

###  3. 기능적 제약 (Syntactic Limitations)

`lambda`는 \*\*하나의 표현식(expression)\*\*만 가질 수 있습니다. 여러 줄의 코드를 작성하거나, 변수 할당(`=`), 반복문(`for`, `while`)과 같은 구문(statement)을 포함할 수 없습니다. 따라서 아주 간단한 기능이 아니면 `lambda`로 구현하는 것 자체가 불가능합니다.

-----

###  그럼에도 `lambda`를 사용하는 이유와 올바른 용도

위와 같은 단점에도 불구하고, `lambda`는 \*\*'간단한 함수를 다른 함수의 인자로 전달'\*\*할 때 매우 유용하며 파이썬다운(Pythonic) 방식으로 여겨집니다.

  * **대표적인 용도**: `sorted`, `map`, `filter` 와 같은 고차 함수(higher-order function)의 인자로 사용될 때 빛을 발합니다.

    ```python
    students = [('Alice', 90), ('Bob', 85), ('Charlie', 95)]

    # 학생 리스트를 '점수' 기준으로 정렬하고 싶을 때
    # 이럴 때 def를 쓰는 것은 오히려 번거롭다.
    sorted_by_score = sorted(students, key=lambda student: student[1])

    print(sorted_by_score)
    # 결과: [('Bob', 85), ('Alice', 90), ('Charlie', 95)]
    ```

    이처럼 한 번만 쓰고 버릴 간단한 함수를 즉석에서 만들어 전달하는 용도로는 `lambda`가 최고의 선택입니다.

`map()` 함수는 지정한 함수를 반복 가능한(iterable) 데이터의 모든 요소에 각각 적용(apply)하여, 그 결과를 묶어서 반환하는 파이썬의 내장 함수입니다. `for` 반복문을 사용하지 않고도 간결하게 동일한 작업을 처리할 수 있게 해줍니다.

-----

###  `map` 함수의 기본 사용법

`map`은 다음과 같은 구조로 사용됩니다.

```python
map(적용할_함수, 함수를_적용할_데이터)
```

  * **적용할 함수 (function)**: 리스트의 각 요소에 적용하고 싶은 함수 이름입니다.
  * **함수를 적용할 데이터 (iterable)**: 리스트, 튜플 등 반복 가능한 데이터를 전달합니다.
  * **반환값**: `map` 함수는 실행 즉시 결과를 계산하지 않고, 필요할 때 값을 생성하는 'map 객체(map object)'를 반환합니다. 따라서 결과를 눈으로 확인하려면 `list()`나 `tuple()`로 변환해야 합니다.

-----

###  예시 1: 모든 숫자를 문자열로 변환하기

숫자로 이루어진 리스트의 모든 요소를 문자열로 바꾸는 예시입니다.

  * **기존 `for` 루프 방식**

    ```python
    numbers = [1, 2, 3, 4, 5]

    print(str_numbers)
    # 결과: ['1', '2', '3', '4', '5']
    ```

    ```python
    numbers = [1, 2, 3, 4, 5]

    # map을 사용하여 int 함수를 numbers의 모든 요소에 적용
    map_object = map(str, numbers)

    # map 객체를 리스트로 변환하여 결과 확인
    str_numbers = list(map_object)

    print(str_numbers)
    # 결과: ['1', '2', '3', '4', '5']
    ```

    `for` 루프를 사용했을 때보다 코드가 훨씬 간결해진 것을 볼 수 있습니다.

`reduce` 함수는 **반복 가능한(iterable) 데이터의 요소를 왼쪽에서 오른쪽으로 하나씩 지정된 함수에 누적 적용하여, 데이터를 단 하나의 값으로 줄여나가는(reduce)** 함수입니다. `functools` 모듈 안에 포함되어 있어 `import` 해서 사용해야 합니다.
- 사실 파이썬 3에서의 사용은 권장되지 않으나, legacy lib 를 쓰는 케이스, 다양한 머신러닝 코드에서 수학 연산 등을 사용시 사용할 수 밖에 없을 수 있다.

-----

###  `reduce` 함수의 기본 사용법

`reduce`는 다음과 같은 구조로 사용됩니다.

```python
from functools import reduce

reduce(누적_적용할_함수, 함수를_적용할_데이터, [초기값])
```

  * **누적 적용할 함수 (function)**: **반드시 두 개의 인자**를 받는 함수여야 합니다. 첫 번째 인자는 누적된 결과(accumulator), 두 번째 인자는 현재 요소(current element)입니다.
  * **함수를 적용할 데이터 (iterable)**: 리스트, 튜플 등 반복 가능한 데이터입니다.
  * **초기값 (initializer, optional)**: 선택적으로 제공하는 초기값입니다. 이 값이 주어지면 첫 번째 연산은 초기값과 리스트의 첫 번째 요소를 가지고 시작합니다. 없으면 리스트의 첫 두 요소로 시작합니다.

-----

###  예시 1: 리스트의 모든 요소 더하기

`reduce`의 가장 대표적인 사용 예시입니다.

  * **기존 `for` 루프 방식**

    ```python
    numbers = [1, 2, 3, 4, 5]
    total = 0
    for n in numbers:
        total += n

    print(total) # 결과: 15
    ```

  * **`reduce` 사용 방식**

    ```python
    from functools import reduce

    numbers = [1, 2, 3, 4, 5]

    # lambda x, y: x + y 는 두 수를 더하는 함수
    # x: 누적값, y: 현재값
    total = reduce(lambda x, y: x + y, numbers)

    print(total) # 결과: 15
    ```

**`reduce`의 동작 과정:**

1.  `x=1`, `y=2` → `1 + 2 = 3` (누적값 3)
2.  `x=3`, `y=3` → `3 + 3 = 6` (누적값 6)
3.  `x=6`, `y=4` → `6 + 4 = 10` (누적값 10)
4.  `x=10`, `y=5` → `10 + 5 = 15` (최종 결과 15)

-----

###  예시 2: 리스트에서 최댓값 찾기

`reduce`를 사용하여 리스트에서 가장 큰 값을 찾을 수도 있습니다.

```python
from functools import reduce

numbers = [3, 5, 2, 8, 4]

# x(누적값)와 y(현재값)를 비교해서 더 큰 값을 반환하는 람다 함수
max_value = reduce(lambda x, y: x if x > y else y, numbers)

print(max_value) # 결과: 8
```

###  `map` vs `reduce`

`map`과 `reduce`는 종종 함께 언급되지만, 목적이 완전히 다릅니다.

  * **`map`**: 각 요소를 **1:1**로 변환하여 **새로운 리스트**를 만듭니다. (입력 개수 = 출력 개수)
  * **`reduce`**: 모든 요소를 하나의 값으로 통합하여 **단 하나의 결과**를 만듭니다. (입력 개수 → 출력 1개)

최근 파이썬에서는 `reduce`의 기능이 `for` 루프로도 충분히 명확하게 표현 가능하고, 때로는 `reduce`가 코드를 더 복잡하게 만든다고 보아 과거만큼 자주 사용되지는 않습니다. 하지만 함수형 프로그래밍 패러다임에서는 여전히 중요한 함수입니다.

# iterable object
- sequence 형 자료형에서 데이터를 순차적으로 추출하는 과정에 대한 객체
- 내부적으로는 `__iter__`, `__next__` 로 구성되어서, iteranble 속성을 가진 객체가 호출이 되면 iterator 라는 다음 값을 전달하는 객체가 동작하는데 그때 그 값들이 next 를 호출한다고 생각하면 됨 
- 딱히 실제 사용 가능성은 거의 없으나, 수동으로 세밀한 제어를 넣을 때, 커스텀 객체 만들시 스스로 이 룰을 수정하여서 세밀한 조정을 할 수 있음.

# generator

`generator`는 한마디로 **'게으른(lazy)' 데이터 생산자**입니다. 모든 데이터를 미리 만들어 창고에 쌓아두는 대신(리스트), 요청이 올 때마다 물건을 **하나씩 즉석에서 만들어주는(yield)** 기계와 같습니다.

이 방식 덕분에 메모리를 획기적으로 절약할 수 있어 대용량 데이터 처리에 필수적입니다.

-----

###  1. Generator란 무엇인가?: `yield` 키워드의 마법

`generator`는 `iterator`를 생성해주는 특별한 함수입니다. 일반 함수와의 가장 큰 차이점은 `return` 대신 `yield` 키워드를 사용하는 것입니다.

  * **일반 함수 (`return`)**: 함수가 호출되면 모든 결과를 **한 번에** 계산하여 리스트 등의 형태로 메모리에 올리고 반환한 뒤 종료됩니다.

    ```python
    def general_list(value):
        result = []
        for i in range(value):
            result.append(i)
        return result # 모든 결과를 담은 리스트를 반환
    ```

  * **제너레이터 함수 (`yield`)**: 함수가 호출되면 즉시 실행되지 않고, `generator` 객체만 반환합니다. 그 후 `for`문 등에서 요청할 때마다 코드를 실행하다가 `yield`를 만나면 값을 **하나만** 반환하고, 그 자리에서 **실행을 일시 정지**합니다. 다음에 다시 요청이 오면 멈췄던 지점부터 실행을 재개합니다.

    ```python
    def generator_list(value):
        for i in range(value):
            yield i # 값을 하나씩 생성하여 반환하고 멈춤
    ```

    이 '생성하고 멈추기' 방식 덕분에 모든 값을 메모리에 저장할 필요가 없어집니다.

-----

###  2. 왜 Generator를 사용하는가?: 메모리 효율성 🧠

슬라이드의 예시는 `generator`의 압도적인 메모리 효율성을 보여줍니다.

  * **리스트 컴프리헨션**: `list_ex = [n*n for n in range(500)]`

      * `range(500)`의 모든 값(500개)에 대해 `n*n`을 미리 계산하고, 그 결과 500개를 전부 메모리에 올려 리스트를 완성합니다.

  * **제너레이터 표현식**: `gen_ex = (n*n for n in range(500))`

      * 이 코드는 계산을 전혀 수행하지 않습니다. 단지 "요청이 오면 `n*n`을 계산할 준비가 되어 있다"는 `generator` 객체 하나만 생성합니다. 이 객체 자체의 크기는 매우 작습니다.

만약 `range(500)`이 아니라 `range(500000000)` 이었다면, 리스트는 엄청난 메모리를 사용하거나 메모리 부족으로 프로그램이 중단될 수 있지만, `generator`는 전혀 문제없이 동작합니다.

-----

###  3. 언제 Generator를 사용하는가?

슬라이드의 내용을 바탕으로 다음과 같은 상황에서 `generator`를 사용하는 것이 좋습니다.

1.  **대용량 데이터 처리**: 처리할 데이터가 너무 커서 메모리에 한 번에 올리기 부담스러울 때 사용합니다. 예를 들어, 수 기가바이트(GB) 크기의 로그 파일을 한 줄씩 읽어 처리하는 경우가 대표적입니다.

2.  **전체 데이터가 필요 없는 경우**: 수백만 개의 데이터 중 특정 조건을 만족하는 첫 번째 데이터만 찾고 싶은 경우, `generator`를 사용하면 해당 데이터를 찾는 즉시 반복을 멈출 수 있습니다. 리스트였다면 필요 없는 뒷부분의 데이터까지 모두 생성하느라 자원을 낭비하게 됩니다.

3.  **무한한 순차 데이터 생성**: 피보나치 수열처럼 끝없이 이어지는 데이터를 만들 때 `generator`를 사용하면, 필요한 만큼만 값을 계속해서 생성해낼 수 있습니다.

4.  **데이터 처리 파이프라인**: 여러 단계의 데이터 처리 과정을 엮을 때 `generator`를 사용하면 매우 효율적입니다. 1단계 `generator`가 파일을 읽고, 2단계 `generator`가 데이터를 정제하며, 3단계 `generator`가 최종 결과를 내는 식으로 데이터를 물 흐르듯 처리할 수 있습니다.

`generator`의 개념과 `itertools.islice` 도구에 대한 내용을 한 페이지로 정리해 드립니다.

### Generator와 `islice` 핵심 정리

**`generator`**는 메모리를 아끼기 위해 데이터를 **필요한 순간에 하나씩 만들어내는 기계**이며, **`islice`**는 이 기계에서 생산된 데이터 스트림을 **원하는 만큼 잘라내는 정교한 도구**입니다.

---
#### 1. Generator: 게으른(Lazy) 데이터 생산자 🏭

* **정의**: `yield` 키워드를 사용한 함수. 호출 시 모든 값을 계산하여 메모리에 저장하는 대신, **요청이 올 때마다 값을 하나씩 생성(`yield`)하고 자신의 상태를 기억하며 일시 정지**합니다.
    

* **핵심 장점**: **압도적인 메모리 효율성**. 데이터가 아무리 커도(수억 개 이상), 단 하나의 값을 처리할 메모리만 있으면 되므로 대용량 데이터를 다루는 데 필수적입니다.

* **주요 사용처**:
    * **대용량 데이터 처리**: 메모리에 담을 수 없는 큰 파일(로그, CSV 등)을 한 줄씩 처리할 때.
    * **검색 및 유효성 검사**: 전체 데이터 중 조건을 만족하는 첫 번째 항목만 찾으면 될 때. 불필요한 뒷부분의 연산을 모두 생략하여 성능을 높입니다.
    * **데이터 파이프라인**: 여러 데이터 처리 단계를 엮을 때, 중간 결과를 메모리에 저장하지 않고 물 흐르듯 데이터를 전달할 수 있습니다.

---
#### 2. `itertools.islice`: 정교한 스트림 절단기 ✂️

* **정의**: `itertools` 모듈의 함수로, 제너레이터와 같은 이터레이터(데이터 스트림)를 **원하는 만큼, 원하는 구간만 잘라내는(slice)** 역할을 합니다.

* **특징**: 제너레이터의 '게으른' 특성을 해치지 않으면서 동작합니다. 즉, 잘라낼 부분까지만 데이터를 요청하여 처리하므로 매우 효율적입니다.

* **주요 기능**:
    * `islice(stream, N)`: 스트림의 앞에서부터 N개만 가져오기.
    * `islice(stream, start, stop)`: `start` 지점까지는 건너뛰고, `stop` 지점까지만 가져오기.

---
### 그래서, 언제 무엇을 쓸까?

두 접근법은 상호 보완적이며, 상황에 따라 더 적합한 방식이 있습니다.

* **직접 '묶음(Chunk)'을 만드는 Generator를 쓸 때**:
    * **상황**: "데이터를 100개씩 묶어서 처리"하는 작업처럼, **'N개짜리 묶음'이 작업의 기본 단위**일 때.
    * **이유**: 코드가 `for chunk in chunk_generator:` 형태로 작성되어, "아, chunk는 100개짜리 리스트구나"라고 이해하기 쉬워 **가독성**이 좋습니다.

* **`islice`를 쓸 때**:
    * **상황**: 이미 "데이터를 한 개씩" 반환하는 기본 제너레이터가 있을 때, 이를 **유연하게 활용**하고 싶을 때.
    * **이유**: "앞에서 10개만 테스트용으로 보자", "1000번째부터 1050번째 데이터만 필요하다" 등 **복잡하고 다양한 방식의 제어**가 필요할 때 훨씬 강력하고 적합합니다.

파이썬 함수에서 인자를 전달하는 다양한 방식들을 정리해 드리겠습니다. 각각 **이름으로 값을 지정**하거나, **기본값을 설정**하거나, **개수가 정해지지 않은 인자**를 유연하게 처리하는 역할을 합니다.

-----

### 1. Default Arguments (기본값 인자)

**정의**: 함수를 **정의할 때** 매개변수(parameter)에 미리 기본값을 할당해두는 방식입니다.

  * **목적**: 함수를 호출할 때 특정 인자값이 들어오지 않으면, 설정된 기본값을 사용하게 하여 **함수 호출을 유연하고 간결하게** 만듭니다. 자주 사용되는 값을 기본값으로 설정해두면 편리합니다.
  * **문법**: `def 함수이름(param1, param2='default_value'):`
  * **주의**: 기본값 인자는 반드시 일반 인자(기본값이 없는 인자) **뒤에 위치**해야 합니다.

<!-- end list -->

```python
# message 인자에 기본값 'Hello'를 설정
def greet(name, message='Hello'):
    print(f"{message}, {name}!")

greet('Alice')         # message 인자를 생략 -> 기본값 'Hello' 사용
# 결과: Hello, Alice!

greet('Bob', 'Hi')     # message 인자를 전달 -> 전달된 값 'Hi' 사용
# 결과: Hi, Bob!
```

-----

### 2. Keyword Arguments (키워드 인자)

**정의**: 함수를 **호출할 때** 매개변수의 이름을 직접 지정하여 값을 전달하는 방식입니다.

  * **목적**:
    1.  **가독성 향상**: `create_user('John', 30, True)` 보다 `create_user(name='John', age=30, active=True)` 가 각 인자의 의미를 훨씬 명확하게 보여줍니다.
    2.  **순서 독립성**: 인자의 순서를 지키지 않고 자유롭게 전달할 수 있습니다.
  * **문법**: `함수이름(param2=value2, param1=value1)`

<!-- end list -->

```python
def create_user(name, age, is_active):
    print(f"Name: {name}, Age: {age}, Active: {is_active}")

# 키워드 인자를 사용하면 순서가 바뀌어도 정상 동작
create_user(age=30, is_active=True, name='Alice')
# 결과: Name: Alice, Age: 30, Active: True
```

-----

### 3. Variable-length Arguments (가변 인자) 📦

**정의**: 함수를 호출할 때 전달되는 **인자의 개수가 정해져 있지 않은 경우**에 사용합니다. `*args`와 `**kwargs` 두 종류가 있습니다.

#### A) `*args` (Positional)

`*`는 여러 개의 \*\*위치 기반 인자(positional arguments)\*\*를 하나의 \*\*튜플(tuple)\*\*로 묶어서 함수 내부로 전달합니다.

```python
# numbers 앞에 *를 붙여 여러 개의 숫자를 튜플로 받음
def sum_all(*numbers):
    print(f"받은 인자들(튜플): {numbers}")
    return sum(numbers)

print(sum_all(1, 2, 3))
# 받은 인자들(튜플): (1, 2, 3)
# 결과: 6

print(sum_all(10, 20, 30, 40, 50))
# 받은 인자들(튜플): (10, 20, 30, 40, 50)
# 결과: 150
```

#### B) `**kwargs` (Keyword)

`**`는 여러 개의 \*\*키워드 인자(keyword arguments)\*\*를 하나의 \*\*딕셔너리(dictionary)\*\*로 묶어서 함수 내부로 전달합니다.

```python
# details 앞에 **를 붙여 여러 키워드 인자를 딕셔셔너리로 받음
def user_profile(name, **details):
    profile = f"Name: {name}\n"
    for key, value in details.items():
        profile += f"- {key}: {value}\n"
    return profile

# age, city, status가 details 딕셔너리로 묶여 전달됨
profile_info = user_profile('Alice', age=30, city='Seoul', status='active')
print(profile_info)
# 결과:
# Name: Alice
# - age: 30
# - city: Seoul
# - status: active
```
