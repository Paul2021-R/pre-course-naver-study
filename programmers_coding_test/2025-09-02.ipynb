```python
def solution(n, w, num):
    # ìµœì´ˆ ì„¤ì •
    total_height = int(n / w)

    if total_height * w < n:
        total_height = total_height + 1

    # ì‹œë®¬ë ˆì´ì…˜ì„ ìœ„í•œ ë°•ìŠ¤ ì„¤ì •
    box_pos = [[0 for _ in range(w)] for _ in range(total_height)]

    # ì¸µê³ ì— ë”°ë¥¸ ë°•ìŠ¤ ë²ˆí˜¸ 
    delimiter_value = 1
    # íƒ€ê²Ÿì´ ë˜ëŠ” ë°•ìŠ¤ ìœ„ì¹˜ ì¸¡ì •ì„ ìœ„í•œ height, width 
    target_height = 1
    target_width = 0

    # ì£¼ìš” ì‹œë®¬ë ˆì´ì…˜(ë°•ìŠ¤ ì ì¬ ë° ì¡°ê±´ì— ë§ëŠ” ìœ„ì¹˜ ë°œê²¬ ë° ì„¸ë¶€ ì„¤ì •
    for index in range(total_height):    
        for x in range(w):
            box_pos[index][x] = delimiter_value + x
            # ì°¾ì„ ë°•ìŠ¤ ìœ„ì¹˜ ì¼ ë•Œ í•  ì¼
            if box_pos[index][x] == num:
                target_height = index
                if (index - 1) % 2 == 0:
                    target_width = w - 1 - x
                else: 
                    target_width = x 
            # ë§ˆì§€ë§‰ ë°•ìŠ¤ê¹Œì§€ ë§ˆë¬´ë¦¬ ì‹œ ë£¨í”„ ê°•ì œ ì¢…ë£Œ
            if box_pos[index][x] == n:
                break
        delimiter_value += w

        # ë°°ì—´ì„ ë’¤ì§‘ì–´ì•¼ í•˜ëŠ” ê²½ìš° íƒìƒ‰
        if (index - 1) % 2 == 0:
            box_pos[index].reverse()


    answer = 0
    for next_height in range(target_height + 1, total_height):
        # ë°•ìŠ¤ê°€ ì¢Œí‘œìƒ ì¡´ì¬í•˜ë©´ 
        if box_pos[next_height][target_width] != 0:
            answer += 1
        else :
            break
    
    return answer + 1
```
